```python
from sentence_transformers import SentenceTransformer, util

# Load model (first time may take a bit of time)
model = SentenceTransformer('all-MiniLM-L6-v2')

# Sample user bio/skills (you can replace this with input())
user_bio = "I am a Python developer with experience in data analysis and machine learning."

# Sample job descriptions
jobs = {
    "Data Scientist": "Looking for a data scientist skilled in Python, machine learning, and data visualization.",
    "Web Developer": "Need a frontend developer with HTML, CSS, JavaScript skills.",
    "AI Engineer": "Looking for someone experienced in deep learning, transformers, and NLP using Python.",
    "Graphic Designer": "Seeking a creative designer proficient in Adobe Photoshop and Illustrator.",
    "Data Analyst": "Require someone skilled in Excel, SQL, Python, and data visualization tools like Power BI."
}

# Convert bio and job descriptions to embeddings
user_embedding = model.encode(user_bio, convert_to_tensor=True)
job_embeddings = {job: model.encode(desc, convert_to_tensor=True) for job, desc in jobs.items()}

# Compute similarity
similarities = {job: util.pytorch_cos_sim(user_embedding, emb).item() for job, emb in job_embeddings.items()}

# Sort and print best match
best_match = max(similarities, key=similarities.get)
print(f"\n👤 Your Bio:\n{user_bio}\n")
print("🧠 AI Best Job Match:")
print(f"✅ {best_match} - {jobs[best_match]}")
print(f"\n🧪 Similarity Score: {similarities[best_match]:.2f}")

# Optional: Print ranked list
print("\n📊 All Job Similarities:")
for job, score in sorted(similarities.items(), key=lambda x: x[1], reverse=True):
    print(f"{job}: {score:.2f}")
```
